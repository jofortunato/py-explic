# Sqlite3

o Sqlite3 é uma versão *light* de SQL que vem integrada como uma biblioteca standard de Python. SQL é uma linguagem de programação muito util para trabalharmos com bases de dados.

Uma base de dados não é nada mais que um conjunto de tabelas com valores.

Para utilizar começamos por importar a biblioteca:

```python
import sqlite3

def read_file(file):
    player_list = []
    
    f = open(file, "r")
    lines = f.readlines()
    
    for line in lines:
        aux_list = line.split(",")
        
        aux_dict = {}
        aux_dict["name"] = aux_list[0]
        aux_dict["age"] = int(aux_list[1])
        aux_dict["total_min"] = int(aux_list[7])
        aux_dict["nacionality"] = aux_list[10]
        aux_dict["total_games"] = int(aux_list[11])
        aux_dict["total_goals"] = int(aux_list[14])
        
        player_list.append(aux_dict)
    f.close()
    return player_list

player_list = read_file("EPL1819.csv")
```

Criar uma conecção e definir um cursor:

```python
conn = sqlite3.connect("players.db") # Criar a ligação com uma base de dados
c = conn.cursor() # Criar o cursor
```

Quando criamos uma **conecção** com uma base de dados, **se a base de dados não existir esta é criada automaticamente**.

Um **cursor** é equivalente ao "cursor do rato", dado que é aquilo que nos **vai permitir fazer diferentes acções**.

Criar uma tabela:

```python
c.execute("CREATE TABLE IF NOT EXISTS Players(name TEXT, age INTEGER, total_min INTEGER, nacionality TEXT,\
total_games INTEGER, total_goals INTEGER)")
```

Tipos de dados em SQL:

    - INTEGER: número inteiro;
    - REAL: número real;
    - TEXT: texto.

Adicionar entradas numa tabela - **INSERT INTO**:

```python
#c.execute("INSERT INTO TestTable VALUES(101,'2020-05-29 13:53:39','Python',6.1)")
```

*.commit()* é equivalente a guardarmos as alterações que fizemos.

Introduzir variáveis numa tabela:

```python
for player in player_list:
    
    c.execute("INSERT INTO Players (name, age, total_min, nacionality, total_games, total_goals)\
    VALUES (?, ?, ?, ?, ?, ?)",
              (player["name"], player["age"], player["total_min"], player["nacionality"], player["total_games"], player["total_goals"]))
```

```python
conn.commit()
```

Ler conteúdo de uma tabela - **SELECT**:

```python
c.execute('SELECT * FROM Players')
data = c.fetchall()

print(data)
```

```python
for row in data:
    print(row)
```

O operador **SELECT** é seguido das colunas da tabela que queremos retornar, por exemplo:

    - * devolve todas as colunas;
    - name devolve apenas essa coluna;
    - name, total_goals devolve a coluna id e a coluna words.

```python
c.execute('SELECT total_goals, name FROM Players') # A ordem com que colocamos as colunas tem inflencia no resultado
data = c.fetchall()

print(data)
```

Temos de utilizar sempre o *.fetchall()* após a pesquisa para recebermos a lista dos resutados.

Podemos adicionar alguma lógica às nossas pesquisas utilizando o operador **WHERE**:

```python
c.execute('SELECT * FROM Players WHERE total_goals > 20')
data = c.fetchall()

print(data)
```

Podemos utilizar o operador **LIKE** para pesquisar termos semelhantes a um determidado valor:

```python
c.execute('SELECT * FROM Players WHERE name LIKE "%David%" ORDER BY total_goals;')
data = c.fetchall()

print(data)
```

```python
c.execute('SELECT DISTINCT nacionality FROM Players')
data = c.fetchall()

print(data)
```

```python
c.execute('SELECT name, max(total_games) FROM Players WHERE total_goals>20;')
data = c.fetchall()

print(data)
```
